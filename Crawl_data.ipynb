```python
import requests
from bs4 import BeautifulSoup

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

def get_article_links(base_url):
    response = requests.get(base_url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        articles = soup.find_all("article", class_="item-news")
        links = [article.find("a")["href"] for article in articles if article.find("a")]
        return links
    else:
        print(f"Không thể tải danh sách bài báo, mã lỗi: {response.status_code}")
        return []

def get_article_content(url):
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        paragraphs = soup.find_all("p", class_="Normal")
        unique_texts = set(p.text.strip() for p in paragraphs)
        content = []
        for text in sorted(unique_texts):
            sentences = text.replace(". ", ".\n").split("\n")
            # Chỉ giữ lại các câu có từ 4 từ trở lên
            filtered_sentences = [sentence.replace('"', '') for sentence in sentences if len(sentence.split()) >= 4]
            content.extend(filtered_sentences)
        return "\n".join(content)
    else:
        print(f"Không thể tải bài báo {url}, mã lỗi: {response.status_code}")
        return ""

def save_to_file(content, filename="news.txt"):
    with open(filename, "a+", encoding="utf-8") as file:
        file.write(content + "\n")
    print(f"Đã thêm nội dung vào: {filename}")

base_url = "https://vnexpress.net/so-hoa/cong-nghe"

if __name__ == "__main__":
    links = get_article_links(base_url)
    if links:
        for i, link in enumerate(links):
            print(f"Đang xử lý bài báo {i + 1}/{len(links)}: {link}")
            content = get_article_content(link)
            if content:
                save_to_file(content)
    else:
        print("Không tìm thấy link bài báo để xử lý.")
```
